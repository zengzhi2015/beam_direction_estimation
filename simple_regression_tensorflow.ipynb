```python
X_data = np.vstack([dataset['MU_X'],dataset['MU_Y'],dataset['SIGMA_XX'],dataset['SIGMA_XY'],dataset['SIGMA_YY']]).T
print(X_data.shape)
```

```python
# Y_data = np.squeeze(dataset['Y_XPOS'])
Y_data = np.vstack([dataset['Y_THETA'],dataset['Y_XPOS']]).T
print(Y_data.shape)
```

# Build network

```python
from keras.layers import Dense
from keras.models import Model
from keras.layers.core import Dropout
from keras.models import Sequential
from keras.layers import BatchNormalization
```

```python
model = Sequential([
    BatchNormalization(input_shape=(5,)),
    Dense(16,activation='tanh'),
    BatchNormalization(),
    Dense(16,activation='tanh'),
    BatchNormalization(),
    Dense(16,activation='tanh'),
    BatchNormalization(),
    Dense(16,activation='tanh'),
    BatchNormalization(),
    Dense(16,activation='tanh'),
    BatchNormalization(),
    Dense(16,activation='tanh'),
    Dense(2)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_absolute_error')
```

```python
training_history = model.fit(X_data, Y_data, batch_size=256, epochs=500, validation_split=0.02)
```

```python
%matplotlib --list
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Plot the history of training
fig = plt.figure() #figsize=(8,6)
plt.plot(training_history.history['loss'][1:])
plt.plot(training_history.history['val_loss'][1:])
plt.title('training loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc='upper left')
plt.savefig('record_data_training_2.png',dpi=600)
plt.show()
```

```python
model.save('Best_5P_DNN.h')
```
